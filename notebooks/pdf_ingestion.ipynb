```python
import feedparser
import requests
import time
import os
```

```python
# !pip install feedparser
```

```python
CATEGORY = "cs.LG"
MAX_RESULTS = 10          # arXiv allows up to 30k with multiple calls
OUTPUT_DIR = "arxiv_papers"

os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
for start in range(0, MAX_RESULTS, 100):
    url = f"http://export.arxiv.org/api/query?search_query=cat:{CATEGORY}&start={start}&max_results=100"
    print("Fetching:", url)

    data = feedparser.parse(url)

    for entry in data.entries:
        pdf_url = entry.id.replace("abs", "pdf") + ".pdf"
        title = entry.title.replace(" ", "_").replace("/", "_")
        filename = f"{OUTPUT_DIR}/{title}.pdf"

        print("Downloading:", filename)
        r = requests.get(pdf_url)
        open(filename, "wb").write(r.content)

    time.sleep(3)     
```
